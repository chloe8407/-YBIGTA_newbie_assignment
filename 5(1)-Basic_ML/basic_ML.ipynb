```python
df_fraud = df[df["Class"] == 1]
df_normal = df[df["Class"] == 0]

df_normal_sample = df_normal.sample(n=10000, random_state=42)

df_sample = pd.concat([df_fraud, df_normal_sample], axis=0).reset_index(drop=True)

df_sample["Class"].value_counts(normalize=True)*100
```

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

df_sample["Amount_Scaled"] = scaler.fit_transform(
    df_sample[["Amount"]]
)

df_sample = df_sample.drop(columns=["Amount"])

X = df_sample.drop(columns=["Class"])
y = df_sample["Class"]
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
y_test.value_counts(normalize=True)*100
```

```python
y_train.value_counts(normalize=True)*100
```

```python
from imblearn.over_sampling import SMOTE

print("SMOTE 적용 전\n", y_train.value_counts())

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

print("SMOTE 적용 후\n", y_train_smote.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, average_precision_score

model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_smote, y_train_smote)

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

print(y_pred)
print(y_proba)

print(classification_report(y_test, y_pred, digits=4))

pr_auc = average_precision_score(y_test, y_proba)
print(pr_auc)
```

```python
import numpy as np

threshold = 0.6
y_pred_thresh = (y_proba >= threshold).astype(int)
```

```python
print(classification_report(y_test, y_pred_thresh, digits=4))

pr_auc = average_precision_score(y_test, y_proba)
print(pr_auc)
```
